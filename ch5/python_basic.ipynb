```python
## 이미지 픽셀 접근하는 방법
img = cv2.imread('villa.jpg')
print(img[100,200])
print(img[100][200])

print()

print(img[100,200,0])
print(img[100][200][0])
print(img.item(100,200,0))
```

```python
img = cv2.imread('villa.jpg')

img = img.astype('float64')
cv2.normalize(img, img, 0, 1, cv2.NORM_MINMAX)
print(img.dtype)
print(img)
# img = img.astype('uint8')

## matplotlib 사용 (RGB 순서)
plt.imshow(img[...,::-1])
plt.show()

## opencv 사용 (BGR 순서)
cv2.imshow('image', img)
cv2.waitKey()
cv2.destroyAllWindows()
```
