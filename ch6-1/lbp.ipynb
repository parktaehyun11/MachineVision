# 2. Application  
## 1) lbp 기술자

```python
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import matplotlib.pyplot as plt
%matplotlib inline

radius = 1
numPoints = 8 * radius
method = 'default'

img1 = cv2.imread('landscape.jpg')
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

gray2 = gray1.copy()
gray2 = gray2 - 100

lbp1 = local_binary_pattern(gray1, numPoints, radius, method)
lbp2 = local_binary_pattern(gray2, numPoints, radius, method)
n_bins = int(lbp1.max()) + 1

plt.subplot(2,1,1)
plt.hist(lbp1.ravel(), bins=n_bins, range=(0,n_bins))
plt.subplot(2,1,2)
plt.hist(lbp2.ravel(), bins=n_bins, range=(0,n_bins))
plt.show()

cv2.imshow('gray', gray1)
cv2.imshow('gray - 100', gray2)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

## 2) texture matching

```python
radius = 1
numPoints = 8 * radius
method = 'uniform'
```

```python
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,8))

## for carpet image
img_carpet = cv2.imread('carpet.jpg')
gray_carpet = cv2.cvtColor(img_carpet, cv2.COLOR_BGR2GRAY)
lbp_carpet = local_binary_pattern(gray_carpet, numPoints, radius, method)
n_bins = int(lbp_carpet.max())+1


## for brick image
img_brick = cv2.imread('brick.jpg')
gray_brick = cv2.cvtColor(img_brick, cv2.COLOR_BGR2GRAY)
lbp_brick = local_binary_pattern(gray_brick, numPoints, radius, method)


## for grass image
img_grass = cv2.imread('grass.jpg')
gray_grass = cv2.cvtColor(img_grass, cv2.COLOR_BGR2GRAY)
lbp_grass = local_binary_pattern(gray_grass, numPoints, radius, method)


## for test1 image
img_test1 = cv2.imread('test1.jpg')
gray_test1 = cv2.cvtColor(img_test1, cv2.COLOR_BGR2GRAY)
lbp_test1 = local_binary_pattern(gray_test1, numPoints, radius, method)
n_bins = int(lbp_test1.max())+1


## for test2 image
img_test2 = cv2.imread('test2.jpg')
gray_test2 = cv2.cvtColor(img_test2, cv2.COLOR_BGR2GRAY)
lbp_test2 = local_binary_pattern(gray_test2, numPoints, radius, method)


## for test3 image
img_test3 = cv2.imread('test3.jpg')
gray_test3 = cv2.cvtColor(img_test3, cv2.COLOR_BGR2GRAY)
lbp_test3 = local_binary_pattern(gray_test3, numPoints, radius, method)

## 결과 확인하기
print('gray_carpet.shape = {}'.format(gray_carpet.shape))
print('gray_brick.shape = {}'.format(gray_brick.shape))
print('gray_grass.shape = {}'.format(gray_grass.shape))
print('gray_test1.shape = {}'.format(gray_test1.shape))
print('gray_test2.shape = {}'.format(gray_test2.shape))
print('gray_test3.shape = {}'.format(gray_test3.shape))

plt.subplot(3,4,1)
plt.imshow(cv2.cvtColor(img_carpet, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('carpet')

plt.subplot(3,4,2)
plt.hist(lbp_carpet.ravel(), bins=n_bins, range=(0,n_bins))

plt.subplot(3,4,3)
plt.imshow(cv2.cvtColor(img_test1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('test1')

plt.subplot(3,4,4)
plt.hist(lbp_test1.ravel(), bins=n_bins, range=(0,n_bins))

plt.subplot(3,4,5)
plt.imshow(cv2.cvtColor(img_brick, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('brick')

plt.subplot(3,4,6)
plt.hist(lbp_brick.ravel(), bins=n_bins, range=(0,n_bins))

plt.subplot(3,4,7)
plt.imshow(cv2.cvtColor(img_test2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('test2')

plt.subplot(3,4,8)
plt.hist(lbp_test2.ravel(), bins=n_bins, range=(0,n_bins))

plt.subplot(3,4,9)
plt.imshow(cv2.cvtColor(img_grass, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('grass')

plt.subplot(3,4,10)
plt.hist(lbp_grass.ravel(), bins=n_bins, range=(0,n_bins))

plt.subplot(3,4,11)
plt.imshow(cv2.cvtColor(img_test3, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('test3')

plt.subplot(3,4,12)
plt.hist(lbp_test3.ravel(), bins=n_bins, range=(0,n_bins))


plt.tight_layout()
plt.show()
```

```python
## 정규화된 lbp 히스토그램 생성하기
hist_carpet, _ = np.histogram(lbp_carpet, bins=n_bins, range=(0, n_bins), density=True)
hist_brick, _ = np.histogram(lbp_brick, bins=n_bins, range=(0, n_bins), density=True)
hist_grass, _ = np.histogram(lbp_grass, bins=n_bins, range=(0, n_bins), density=True)
hist_test1, _ = np.histogram(lbp_test1, bins=n_bins, range=(0, n_bins), density=True)
hist_test2, _ = np.histogram(lbp_test2, bins=n_bins, range=(0, n_bins), density=True)
hist_test3, _ = np.histogram(lbp_test3, bins=n_bins, range=(0, n_bins), density=True)
```
